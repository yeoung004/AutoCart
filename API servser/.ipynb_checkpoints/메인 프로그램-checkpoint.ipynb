### 자동장바구니 주기(하루)

```python
def autoBasket(basket_id_count,basket_id_unique,cursor,pcode):
    try:
        # 사용자 별로 로그인
        count = 0
        #여기서부터 넣어야 될 회원수 자르기
        url="https://checkout.iherb.com/users/dashboard"
        driver = wb.Chrome()
        driver.get(url)
        #아이디, 비밀번호 가져오기
        cursor.execute("""select IHERBID,IHERBPW from mbr where id = '{}' """.format(basket_id_unique))
        iherb = []
        for index in cursor:
            iherb = index

        iherbId = driver.find_element_by_id("username_input")
        iherbPw = driver.find_element_by_id("Password")

        iherbId.send_keys(iherb[0])
        iherbPw.send_keys(iherb[1])
        iherbPw.send_keys(Keys.ENTER)
        #여기까지 로그인

        # id별로 로그인-장바구니 담기를 해줌
        btn = driver.find_element_by_class_name("js-add-more-quickshop")
        btn.click()
        btn.click()

        for j in range(basket_id_count):
            #담을 상품 개수
            baseket = driver.find_element_by_id('product-code-{}'.format(j))
            baseket.send_keys(pcode[j+count])
            time.sleep(3)

        count = basket_id_count

        btn = driver.find_element_by_class_name("quickshop-submit")
        btn.click()
        url = 'https://secure.iherb.com/account/logoff/'
        driver.get(url)
#여기까지 장바구니 넣기
        driver.quit()
        
    except Exception as ex:
        print("오류 발생" + str(ex))
        print(traceback.format_exc())
        driver.quit()
```

```python
def thread_ind(basket_id_count,basket_id_unique,cursor,pcode):
    #여기에 코드넣기
    #하루에 한번씩 실행 216000
    thread = threading.Thread(target=autoBasket(basket_id_count,basket_id_unique,cursor,pcode))
    thread.daemon = True
    thread.start()
```

### 분석

```python
def analysis(pcode):
    #리뷰 데이터를 저장한 CSV파일 불러오기
    rating_data = pd.read_csv('findX.csv')
    
    #상품 데이터 정보들 불러오기
    product_data = pd.read_csv('제품상세.csv')
    
    #Data Preprocessing
    product_data = product_data.drop(['price','star','quantity','company','url','img_url'],axis='columns')
    user_product_data = pd.merge(rating_data, product_data, on = 'pcode', )
    user_product_rating = user_product_data.pivot_table('rate', index = 'Id', columns='pcode').fillna(0)
    product_user_rating = user_product_rating.values.T
    SVD = TruncatedSVD(n_components=10)
    matrix = SVD.fit_transform(product_user_rating)
    corr = np.corrcoef(matrix)
    product_title = user_product_rating.columns
    product_title_list = list(product_title)
    coffey_hands = product_title_list.index(pcode)
    corr_coffey_hands = corr[coffey_hands]
    result = list(product_title[(corr_coffey_hands >= 0.6)])[:4]
    result.remove(pcode)
    return result
```

## 리뷰 가져오기

```python
def getReview(userId):
    try:
        #리뷰 정보 가져오는 매크로를 실행 하기위해 크롬 웹 드라이버 실행
        url="https://www.iherb.com/ugc/myaccount/review?&filter=3"
        driver = wb.Chrome()
        driver.get(url)
        
        #회원가입후 톰켓에서 넘겨받아온 아이디에 해당하는 쇼핑몰 아이디, 비밀번호 정보 가져오기
        cursor.execute("""select * from mbr where id='{}'""".format(userId))
        list1 = cursor.fetchone()
        
        # 쇼핑몰 아이디, 비밀번호 정보 배열에 저장 및 쇼핑몰 로그인
        user_id = list1[2]
        user_pw = list1[3]
        iherbId = driver.find_element_by_id("username_input")
        iherbPw = driver.find_element_by_id("Password")
        
        #유저들의 리뷰목록 가져오는 Macro --start
        try:
            iherbId.send_keys(user_id)
            iherbPw.send_keys(user_pw)
            iherbPw.send_keys(Keys.ENTER)
            time.sleep(2)
            iherbId.send_keys(user_id)
            check = False
        except:
            check = True
            
        star_list = []
        error_ch = False
        score = 0
        score_list = []
        code = []

        # 다음 페이지 버튼
        more_btn = driver.find_elements_by_css_selector("div.css-rmfzgn > button")

        for page in range(1,len(more_btn)-1):
            more_btn = driver.find_elements_by_css_selector("div.css-rmfzgn > button")
            more_btn[page].click()
            star_list = driver.find_elements_by_css_selector("ul.css-1mokw6q > li.css-172co2l path")
            pcode_list = driver.find_elements_by_css_selector("div.css-9u0z6w > a > img")
            for i in range(len(pcode_list)):
                code.append(pcode_list[i].get_attribute("src").split('/')[4])
                
            try:
                for i in range(0,len(star_list),5):
                    for j in range(i,(i+5)):
                        if star_list[j].get_attribute("fill") == "#FAC627":
                            score = score + 1
                        else:
                            break
                    score_list.append(score)
                    score = 0
            except:
                print("에러발생")
                continue


        id_list = []
        for i in range(len(score_list)):
            id_list.append(userId)
        
        for i in range(len(code)):
            code[i] = code[i][:3].upper() + '-' + code[i][3:]
        #유저들의 리뷰목록 가져오는 Macro --end
        
        #머신러닝을 사용하기 위해 가져온 정보들을 findX.CSV로 저장한다.
        findX = pd.read_csv('findX.csv')    
        csv_dic = {'Id':id_list,'pcode':code,'rate':score_list}
        review = pd.DataFrame(csv_dic)
        review = pd.concat([review,findX])
        review.set_index('Id',inplace=True)
        review.to_csv('findX.csv')
        driver.quit()
    except Exception  as ex:
        print("리뷰 에러 발생" + str(ex))
        driver.quit()
```

### 메신저

```python
import sys
from sdk.api.message import Message
from sdk.exceptions import CoolsmsException

##  @brief This sample code demonstrate how to send sms through CoolSMS Rest API PHP
def message(userId):
    # set api key, api secret
    api_key = "#ENTER_YOUR_OWN#"
    api_secret = "#ENTER_YOUR_OWN#"

    ## 4 params(to, from, type, text) are mandatory. must be filled
    params = dict()
    params['type'] = 'sms' # Message type ( sms, lms, mms, ata )
    params['to'] = '01000000000' # Recipients Number '01000000000,01000000001'
    params['from'] = '01000000000' # Sender number
    params['text'] = '%s님의 장바구니에 상품이 담겼습니다.'% userID # Message

    cool = Message(api_key, api_secret)
    try:
        response = cool.send(params)
        print("Success Count : %s" % response['success_count'])
        print("Error Count : %s" % response['error_count'])
        print("Group ID : %s" % response['group_id'])

        if "error_list" in response:
            print("Error List : %s" % response['error_list'])

    except CoolsmsException as e:
        print("Error Code : %s" % e.code)
        print("Error Message : %s" % e.msg)

    sys.exit()
```
